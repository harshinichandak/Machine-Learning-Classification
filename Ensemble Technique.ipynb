```python
et.isnull().sum()
```

```python
et.drop('Unnamed: 32',axis=1, inplace=True)
et.drop('id',axis=1, inplace=True)
et.shape
```

```python
et.dtypes
```

```python
et.diagnosis.value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
et.diagnosis = le.fit_transform(et.diagnosis)
et.dtypes
```

```python
from sklearn.model_selection import train_test_split

et_train, et_test = train_test_split(et, test_size = .2, random_state = 123)
et.columns.get_loc('diagnosis')
```

```python
et_train_x = et_train.iloc[:, 1:-1]
et_train_y = et_train.iloc[:, 0]

et_test_x = et_test.iloc[:, 1:-1]
et_test_y = et_test.iloc[:, 0]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
dt = DecisionTreeClassifier()
```

```python
dt.fit(et_train_x, et_train_y)
```

```python
pred_dt = dt.predict(et_test_x)
```

```python
tab_dt = confusion_matrix(et_test_y, pred_dt)
tab_dt
```

```python
tab_dt.diagonal().sum()/tab_dt.sum()
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
abc = AdaBoostClassifier(dt)
abc.fit(et_train_x, et_train_y)
pred_abc = abc.predict(et_test_x)

tab_abc = confusion_matrix(et_test_y, pred_abc)
tab_abc
```

```python
tab_abc.diagonal().sum()/tab_abc.sum()
```
